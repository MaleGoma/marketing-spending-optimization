## Preprocesamiento de datos

### Estilo de encabezado

Dado que se encontró que los nombres de las columnas del dataframe 'visits' y 'orders' no siguen las reglas de buen estilo, se procederá a cambiarlos para que los cumplan.

```python
# Creamos una lista de apoyo donde se colocarán todos los nuevos nombres de las columnas
new_col_names = []

for old_name in visits.columns:
# Eliminamos los espacios al principio y al final
    name_stripped = old_name.strip()
# Pasamos todas las letras en minúsculas
    name_lowered = name_stripped.lower()
# Reemplazamos los espacios entre palabras por guiones bajos
    name_no_spaces = name_lowered.replace(' ', '_')
# Agregamos el nuevo nombre a la lista de nuevos nombres de columna
    new_col_names.append(name_no_spaces)

# Reemplazamos los nombres anteriores con los nuevos
visits.columns = new_col_names

#Verificamos el cambio de nombres de las columnas
print(visits.columns)
```

```python
# Creamos una lista de apoyo donde se colocarán todos los nuevos nombres de las columnas
new_col_names = []

for old_name in orders.columns:
# Eliminamos los espacios al principio y al final
    name_stripped = old_name.strip()
# Pasamos todas las letras en minúsculas
    name_lowered = name_stripped.lower()
# Reemplazamos los espacios entre palabras por guiones bajos
    name_no_spaces = name_lowered.replace(' ', '_')
# Agregamos el nuevo nombre a la lista de nuevos nombres de columna
    new_col_names.append(name_no_spaces)

# Reemplazamos los nombres anteriores con los nuevos
orders.columns = new_col_names

#Verificamos el cambio de nombres de las columnas
print(orders.columns)
```

## Análisis de datos

### Visitas

#### ¿Cuántas personas lo usan cada día, semana y mes?

Para encontrar estos datos, se procederá a encontrar las métricas DAU (usuarios activos diarios), WAU (usuarios activos semanales) y MAU (usuarios activos mensuales).

```python
# Obtenemos el día, semana, mes y año de los inicios de sesión
visits['session_year'] = visits['start_ts'].dt.year
visits['session_month'] = visits['start_ts'].dt.month
visits['session_week'] = visits['start_ts'].dt.isocalendar().week
visits['session_date'] = visits['start_ts'].dt.date

# Creamos nuevas columnas que combinen el año con el mes y la semana
visits['year_week'] = visits['start_ts'].dt.strftime('%Y-%U')
visits['year_month'] = visits['start_ts'].dt.strftime('%Y-%m')
```

```python
visits.head()
```

```python
# Obtenemos el DAU (usuarios activos diarios)
dau_total = visits.groupby('session_date').agg({'uid': 'nunique'}).mean()
int(dau_total)
```

```python
# Obtenemos el WAU (usuarios activos semanales)
wau_total = (
    visits.groupby(['session_year', 'session_week'])
    .agg({'uid': 'nunique'})
    .mean()
)
int(wau_total)
```

```python
# Obtenemos el WAU (usuarios activos semanales) con year_week
wau_total_2 = (
    visits.groupby(['year_week'])
    .agg({'uid': 'nunique'})
    .mean()
)
int(wau_total_2)
```

```python
# Obtenemos el MAU (usuarios activos mensuales)
mau_total = (
    visits.groupby(['session_year', 'session_month'])
    .agg({'uid': 'nunique'})
    .mean()
)
int(mau_total)
```

```python
# Obtenemos el MAU (usuarios activos mensuales)
mau_total_2 = (
    visits.groupby(['year_month'])
    .agg({'uid': 'nunique'})
    .mean()
)
int(mau_total_2)
```

Observamos que la plataforma cuenta con 907 usuarios activos diarios, 5716 usuarios activos semanales y 23,228 usuarios activos mensuales.

```python
# Convertimos session_date a datetime
visits['session_date'] = pd.to_datetime(visits['session_date'])

dau = visits.groupby('session_date').agg({'uid': 'nunique'}).reset_index()

# Hacemos la gráfica de línea
dau.plot( 
    x = 'session_date',
    y = 'uid', 
    kind = 'line',
    title = 'Usuarios Activos Diarios (DAU)', 
    xlabel = "Día",
    ylabel = "Cantidad de usuarios activos",
    figsize = [15, 6])

# Especificamos las leyendas
plt.legend(['Usuarios activos'])
plt.show()
```

Observamos que a partir de septiembre de 2017 el número de usuarios activos empezó a aumentar, llegando a un pico en el día 24 de noviembre de 2017. Destaca también un día donde se tuvo un mínimo de usuarios el 31 de marzo de 2018. Igualmente, se nota una pequeña tendencia a la baja a partir de esta fecha.

```python
wau = visits.groupby('year_week').agg({'uid': 'nunique'}).reset_index()

# Hacemos la gráfica de barras
wau.plot( 
    x = 'year_week',
    y = 'uid', 
    kind = 'line',
    title = 'Usuarios Activos Semanales (WAU)', 
    xlabel = "Año / Semana",
    ylabel = "Cantidad de usuarios activos",
    figsize = [15, 6])

# Especificamos las leyendas
plt.legend(['Usuarios activos'])
plt.show()
```

Observamos una tendencia a la alza en usuarios semanales activos a partir de la semana 31 de 2017, donde solo se experimentó una caída drástica en la semana 53 del mismo año. Posteriormente, se nota una pequeña disminución en usuarios en la semana 13, 17 y 18 de 2019.

```python
mau = visits.groupby('year_month').agg({'uid': 'nunique'}).reset_index()

# Hacemos la gráfica de barras
mau.plot( 
    x = 'year_month',
    y = 'uid', 
    kind = 'line',
    title = 'Usuarios Activos Mensuales (MAU)', 
    xlabel = "Mes",
    ylabel = "Cantidad de usuarios activos",
    figsize = [10, 6])

# Especificamos las leyendas
plt.legend(['Usuarios activos'])
plt.show()
```

Se observa una tendencia a la alza en los usuarios activos mensuales a partir de agosto de 2017, donde se alcanzó un máximo en noviembre, y posteriormente se experimentó una bajada, donde los meses abril y mayo se mantuvieron estables.

#### ¿Cuántas sesiones hay por día? (Un/a usuario/a puede tener más de una sesión).

```python
# Obtenemos el número de sesiones por día
sessions_per_day = visits.groupby('session_date').agg({'uid': 'count'}).mean()
int(sessions_per_day)
```

Observamos que, en promedio, se cuenta con 987 sesiones por día.

```python
sessions_per_days = visits.groupby('session_date').agg({'uid': 'count'}).reset_index()

# Hacemos la gráfica de barras
sessions_per_days.plot( 
    x = 'session_date',
    y = 'uid', 
    kind = 'line',
    title = 'Sesiones por día', 
    xlabel = "Día",
    ylabel = "Cantidad de sesiones",
    figsize = [15, 6])

# Especificamos las leyendas
plt.legend(['Sesiones activas'])
plt.show()
```

Observamos que el número de sesiones por día tiene una tendencia similar a el número de usuarios activos diarios. Resaltamos que el DAU promedio obtenido fue de 907 y las sesiones promedio por día día fueron 987, por lo que observamos que la mayoría de los usuarios no suelen mantener más de una sesión abierta.

#### ¿Cuánto dinero traen? (LTV - Life Time Value)

```python
del orders['buy_ts_month_year']

orders.head()
```

```python
# Obtenemos los meses del dataframe orders

orders['buy_ts_month'] = orders['buy_ts'].astype('datetime64[M]')
orders.head()
```

```python
# Obtenemos el mes de la primera compra de cada cliente

first_orders = orders.groupby('uid').agg({'buy_ts_month': 'min'}).reset_index()
first_orders.columns = ['uid', 'first_order_month']
first_orders.head()
```

```python
# Unimos los meses de la primera compra y la tabla de pedidos

orders_ = pd.merge(orders, first_orders, on='uid')

orders_.head()
```

```python
# Agrupamos por mes de primera compra y mes de compra, y sumamos los ingresos

cohorts = orders_.groupby(['first_order_month','buy_ts_month']).agg({'revenue': 'sum'}).reset_index()
cohorts.columns = ['first_order_month', 'order_month', 'revenue']
cohorts.head()
```

```python
# Calculamos el número de nuevos clientes para cada mes

cohort_sizes = first_orders.groupby('first_order_month').agg({'uid': 'nunique'}).reset_index()
cohort_sizes.columns = ['first_order_month', 'n_buyers']
cohort_sizes.head()
```

```python
# Agregamos info de no. de número de nuevos clientes cada mes a cohorts

report = pd.merge(cohort_sizes, cohorts, on='first_order_month')
report.head()
```

```python
# Obtenemos edad de la cohorte

report['age'] = (
    report['order_month'] - report['first_order_month']
) / np.timedelta64(1, 'M')
report['age'] = report['age'].round().astype('int')

report.head()
```

```python
# Obtenemos LTV

report['ltv'] = report['revenue'] / report['n_buyers']

report.head()
```

```python
output = report.pivot_table(
    index='first_order_month', 
    columns='age', 
    values='ltv', 
    aggfunc='mean'
)

output = output.cumsum(axis=1)

output.fillna('')
```

```python
plt.figure(figsize=(13, 9))
plt.title('LTV por cohorte')
sns.heatmap(
    output,
    annot=True,
    fmt='.2f',
    linewidths=1,
    linecolor='gray'
)

plt.xlabel('Edad de la cohorte')
plt.ylabel('Cohorte')

# Mostrar el gráfico
plt.show()
```

Observamos que cada cliente de la primera cohorte representó un ingreso de 11.88 para la empresa. Destaca la cohorte de septiembre de 2017, pues representó ingresos de 13.44 por cliente.

#### ¿Cuán rentables eran las inversiones? (ROMI)

```python
# Obtenemos la primera visita de cada usuario
first_login = visits.groupby('uid')['start_ts'].min().reset_index()

first_login.head()
```

```python
# Obtenemos la fuente de la primera visita de cada usuario

source_visit = pd.merge(first_login, visits[['uid', 'start_ts', 'source_id']], on=['uid', 'start_ts'], how='left')

source_visit.head()
```

```python
# Obtenemos el mes de la primera compra de cada usuario

first_month_order = orders.groupby('uid')['buy_ts_month'].min().reset_index()
first_month_order.columns = ['uid', 'month_first_purchase']

first_month_order.head()
```

```python
# Unimos datos de orders con fuente de la primera visita de cada usuario

users_orders = pd.merge(orders, source_visit[['uid', 'source_id']], on=['uid'], how='left')

users_orders.head()
```

```python
# Unimos datos de orders con el mes de la primera compra de cada usuario

users_orders = pd.merge(users_orders, first_month_order, on=['uid'], how='left')

users_orders.head()
```

```python
# Obtenemos edad de la cohorte

users_orders['age'] = (users_orders['buy_ts_month'] - users_orders['month_first_purchase']) / np.timedelta64(1, 'M')
users_orders['age'] = users_orders['age'].round().astype('int')

display(users_orders.head())
```

```python
# Obtenemos los meses del dataframe costs

costs['month_first_purchase'] = costs['dt'].astype('datetime64[M]')
costs.head()
```

```python
# Obtenemos lista de source_id
source_ids = costs['source_id'].unique()
source_ids
```

```python
# Iteramos sobre source_ids
for source in source_ids:
    
    # Filtramos la tabla por source
    filtered_data = users_orders[users_orders['source_id']==source]
    
    # Calculamos el tamaño de cada cohorte
    cohort_sizes = filtered_data.groupby('month_first_purchase').agg({'uid': 'nunique'}).reset_index()
    cohort_sizes.columns = ['month_first_purchase', 'n_buyers']
    
    # Agrupamos por mes de primera compra y mes de compra, y sumamos los ingresos
    cohorts = filtered_data.groupby(['month_first_purchase','buy_ts_month']).agg({'revenue': 'sum'}).reset_index()
    cohorts.columns = ['month_first_purchase', 'order_month', 'revenue']
    
    # Agregamos info de tamaño de cohortes y cohorts
    data_source = pd.merge(cohort_sizes, cohorts, on='month_first_purchase')
    
    # Obtenemos edad de la cohorte
    data_source['age'] = (
        data_source['order_month'] - data_source['month_first_purchase']
    ) / np.timedelta64(1, 'M')
    data_source['age'] = data_source['age'].round().astype('int')
    
    # Obtenemos LTV
    data_source['ltv'] = data_source['revenue'] / data_source['n_buyers']
    
    # Obtenemos costos mensuales de source
    
    # Filtamos costos por source
    costs_filtered = costs[costs['source_id']== source]
    
    # Obtenemos costos mensuales
    monthly_costs = costs_filtered.groupby(['month_first_purchase']).agg({'costs': 'sum'}).reset_index()
    monthly_costs.columns = ['month_first_purchase', 'costs']
    
    # Agregamos info de costos a data_source
    data_source = pd.merge(data_source, monthly_costs, on='month_first_purchase')
    
    # Obtenemos CAC
    data_source['cac'] = data_source['costs'] / data_source['n_buyers']
    
    # Obtenemos el ROMI
    data_source['romi'] = data_source['ltv'] / data_source['cac']
    
    # Obtenemos tabla pivote
    output = data_source.pivot_table(
        index='month_first_purchase', 
        columns='age', 
        values='romi', 
        aggfunc='mean')
    output = output.cumsum(axis=1)
    output.round(2).fillna('')
    
    # Obtenemos mapa de calor
    plt.figure(figsize=(13, 9))
    plt.title(f'ROMI por cohorte de la fuente {source}')
    sns.heatmap(
        output,
        annot=True,
        fmt='.2f',
        linewidths=1,
        linecolor='gray')
    plt.xlabel('Edad de la cohorte')
    plt.ylabel('Cohorte')
    
    # Mostrar el gráfico
    plt.show()
    
    #print(data_source.head())
```

Observamos que el ROMI para la fuentes fuentes son:
- Fuente 1:
    - El ROMI es entero para la cohorte de noviembre de 2017 y las más antiguas. Para el resto tenemos aún un ROMI decimal. Por lo que para estas primeras cohortes ya se recuperó el costo de la inversión, pero no para las posteriores.
    - El ROMI para la mayoría de las cohortes fue superior a 1 dentro del primer mes de edad de la cohorte.
- Fuente 2:
    - Solo para la mitad de las cohortes se ha recuperado el costo de la inversión en marketing hasta el momento.
    - Vale la pena analizar a profundidad a la cohorte de mayo 2015, pues es la única que fue mayor a uno antes del primer mes de edad de la cohorte.
- Fuente 3:
    - Todos los ROMI de esta fuente no fueron superiores a 1, por lo que no se está recuperando el costo.
    - Solo la cohorte de septiembre 2017 alcanzó llegar a 0.5 de ROMI.
- Fuente 4:
    - La cohorte que tuvo un valor ROMI superior desde el inicio fue la de noviembre de 2017
    - La otra única cohorte que superó el valor de 1 en el ROMI fue junio 2017.
- Fuente 5:
    - Se observa que la cohorte de septiembre 2017 fue la que mayor retorno sobre la inversión en marketing dio, por lo que valdría la pena analizar el por qué de este dato.
    - Notamos una disminución del ROMI en las cohortes a partir de noviembre de 2017.
